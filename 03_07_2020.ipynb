# Importing Libraries:


```python
import numpy as np
import pandas as pd
```

# Reading the data:


```python
data = pd.read_csv("https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/spam.csv",encoding='latin-1')
data.head()
```

```python
data.shape
```

```python
data.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'],axis=1, inplace=True)
data.shape
```

# Splitting the data:

```python
# split the data
# 80% of the data will be in training data  20% of the data will be in testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(data.v2,data.v1,test_size = 0.2)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

# initialising count vectorizer

```python
# count vectorizer
from sklearn.feature_extraction.text import CountVectorizer

# creating an object for countvectorizer
count_vect = CountVectorizer()
```

# To get the feature names:

```python
# generate the word counts for the words in the documents
word_count_vector = count_vect.fit(x_train)

# to get the feature names
word_count_vector.get_feature_names()
```

# Transforming the data:

```python
doc_array = word_count_vector.transform(x_train)
doc_array
```

```python
# converting to a array
doc_array = word_count_vector.transform(x_train).toarray()
doc_array
```

# Specifying column names

```python
# converting the array into a dataframe
pd.DataFrame(doc_array)
```

```python
feature_matrix = pd.DataFrame(doc_array,columns= word_count_vector.get_feature_names())
feature_matrix
```

```python
# TFIDF Vectorizer
# Import TfIDF Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize and object for the TFIDF Vectorizer
tfidf = TfidfVectorizer()
```

```python
# apply the tfidf to the data(x_train)
x_train_transformed = tfidf.fit_transform(x_train)
x_train_transformed
```

```python
x_test_transformed = tfidf.transform(x_test)
x_test_transformed
```

```python
# Feature Names
tfidf.get_feature_names()
```

# Vocabulary Position:

```python
# position of the words in the sparse matrix
tfidf.vocabulary_
```

```python
# Idf of the terms
tfidf.idf_
```

```python
# Apply the naive Bayes Algorithm
# Import BernNB
from sklearn.naive_bayes import BernoulliNB
# creating an object for BerNB
model_BernNB = BernoulliNB()
```

```python
# Applying the algorithm to the data
# objectName.fit(Input,Output)
model_BernNB.fit(x_train_transformed, y_train)
```

```python
# prediction on test data
# SYNTAX: objectname.predict(InputValues)
y_test_pred = model_BernNB.predict(x_test_transformed)
```

```python
# compare the actual values(y_test) with predicted values(y_test_pred)
from sklearn.metrics import confusion_matrix,classification_report
confusion_matrix(y_test,y_test_pred)
```

```python
print(classification_report(y_test,y_test_pred))
```

```python
# test a single message --> Spam or Ham
new_message = pd.Series('Last few days to collect the free entry passes with 40% discount')
#calculate tfidf values for the new message
new_message_transformed = tfidf.transform(new_message)
```
